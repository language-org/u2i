## What features can predict perceived happiness (across countries)? Why are anglo-saxon the happiest?

### Approach: linear model
- We hypothesize that happiness linearly depends on the features (linear regression model).
- We first check **linear dependency** (matrix rank) between features which will impair weights interpretability (not prediction performance). We also check **correlation**. Features' weights can be disambiguated only if features are linearly independent.

```python
# code
# diagonal correlation matrix
if not 'euDistance' in happyMetrics:
    happyMetrics += ['euDistance']
predictors = selFeatures
utils.plot_diag_corrMatx(dataMean[predictors + happyMetrics], (6,6), t0, print_opt=False, numFig=1)

# matrix rank
rank = np.linalg.matrix_rank(dataMean[predictors])
print("Feature matrix rank is:", rank)
```

**Results**
- All features are mutually linear dependent (rank=0): infinite weight solutions to the model and weights will be unreliable.  
- GDP correlated with Ladder, euclidean distance (r>0.7) and other features: Social 
support, Life expectancy (r>0.7), not so much too affects --> **validates rank analysis**.
- Perceived corruption & negative affects mildly negatively correlated with features and positive affect (r<-0.2) --> **makes sense**.
- Surprisingly, life expectancy completely explains euclidean distance --> **to explore**.

**Conclusion**

- Multicollinearity is a major problem here - solutions ?
    - drop all features except one. But which one?
        - based on a subjective prior (Bayesian-like)
        - based on prediction performance (cross-validated feature selection)
        - aggregating features: 
            - with PCA (uninterpretable features)
            - or ZCA (interpretable, https://martin-thoma.com/zca-whitening/)
        
        
- A simple hypothesis is that only one feature (either GDP per capita, social support or life expectancy: highest r) "drives" felt happiness while other features just correlate with that feature. Which feature ?



### Is the model predictive at all?

```python
# cross - validated goodness of fit
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.linear_model import LinearRegression as linreg
from sklearn.preprocessing import scale
dataMeanClean = dataMean.dropna()
X = scale(dataMeanClean[predictors]) # feature and target z-scoring
y = scale(dataMeanClean["Life Ladder"])
scores = cross_val_score(linreg(), X, y, cv=10) # cv-regression w/o feature elimination
print("10-Fold CV mean/std score:", (scores.mean(), scores.std()))

# plot goodness of fit
fig = plt.figure(figsize=(15, 6))
ax1 = fig.add_subplot(121);
pd.DataFrame(scores).plot(kind="bar", ax=ax1, width=0.9);
ax1.set_xlabel("Folds")
ax1.set_ylabel("Model r-squared (1-SSE/SST)")
ax1.axhline(scores.mean(), linestyle = '--');
ax1.axhline(scores.mean() + scores.std(), linestyle = '--');
ax1.axhline(scores.mean() - scores.std(), linestyle = '--');
ax1.set_title("CV - goodness of fit");

# plot felt happiness vs. model prediction
predAndLadder = pd.DataFrame(cross_val_predict(linreg(), X, y, cv=10), 
                             index = dataMeanClean["Country"], columns = ["Predictions"])
predAndLadder["Ladder"] = y;
ax2 = fig.add_subplot(122);
predAndLadder.plot(ax=ax2);
ax2.set_xlabel("Countries");
ax2.set_ylabel("Felt happiness (z-scored)");
ax2.set_title("Predicted vs. Felt");

# Does model performance statistically differ from chance ?
# chance hypothesis has R-squared = 0
```

**Results**
- Model tracks felt happiness well (80% of explained variance) --> **assumption of linearity is not bad.**
- Not a surprise, expected from the observed correlations.

### Which weights are not  0 ?

```python
# import statsmodels as stats
# help(stats.regression.linear_model)
```

Instead of selecting a model based on adjusted R^22 we cross-validate models.
We could plot the weights std to evaluate how much multicollinearity affects the weights variability

```python
# cross validated best feature selection
from itertools import combinations as comb 
from ipdb import set_trace as st

# features testing 
scoresAll = []
cmbAll = []
nTestFeat = len(predictors)
while not nTestFeat == 0:     
    cmb = list(comb(predictors, nTestFeat))
    scores = [cross_val_score(linreg(), scale(dataMeanClean[list(cmb_i)]), y, cv=10).mean() for cmb_i in cmb]
    scoresAll += scores
    cmbAll += cmb
    nTestFeat = nTestFeat - 1
bestFeats = list(cmbAll[np.argmax(scoresAll)]) # selection (max adjusted-R^2)

# best model
X = scale(dataMeanClean[bestFeats]) 
scores = cross_val_score(linreg(), X, y, cv=10) # cv-regression w/o feature elimination
print("Best features:", bestFeats)
print("R^2 (10-Fold mean/std):", (scores.mean(), scores.std()))

# model parameters
model = linreg().fit(X,y)

# plot
fig = plt.figure()
ax = fig.gca()
pd.DataFrame(model.coef_, index=bestFeats, columns=['weights']).sort_values(by="weights", ascending=False).plot(kind='bar', ax=ax, width=0.9);
ax.set_ylabel('Weights');
```

If we assume GDP causally affects perceived happiness, GDP is its strongest driver: when it increases by one unit of variance, felt happiness increases by 0.36 unit of variance.
